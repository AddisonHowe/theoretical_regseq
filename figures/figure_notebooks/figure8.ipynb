```python
# processing
import numpy as np
import pandas as pd
from Bio import SeqIO

# custom packages
import tregs

# visualization
import matplotlib.pyplot as plt
```

### Loading input data

```python
_genome = []
for record in SeqIO.parse("../../data/mg1655_genome.fasta", "fasta"):
    _genome.append(str(record.seq))
genome = _genome[0]
```

```python
# Promoter
rnap_site = 'CAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG'

# Operator sequence
rep_site = 'AATTGTGAGCGGATAACAATT'

## Find the transcription start site
df = pd.read_csv('../../data/ecoli_gene_list.csv')
lacZYA_TSS = int(df[(df.gene == 'lacZ') & (df.promoter == 'lacZp1')]['tss'].iloc[0])
promoter = tregs.seq_utils.get_regulatory_region(genome, lacZYA_TSS, reverse=True)
promoter
```

```python
# obtained from https://journals.plos.org/ploscompbiol/article/figure?id=10.1371/journal.pcbi.1002811.g002

rnap_emat_raw = np.loadtxt('../../data/energy_matrices/RNAP_energy_matrix.txt')
rnap_emat = tregs.simulate.fix_wt(rnap_emat_raw, rnap_site)
```

```python
rep1_site = promoter[(115):(115+10)]
rep2_site = promoter[(115+20):(115+30)]

#promoter = promoter.replace(rep2_site, 'G'+ rep1_site[1:])
#rep2_site = 'G' + rep1_site[1:]

r1_emat = tregs.simulate.generate_emap(rep1_site, fixed=True, fixed_value=1)
r2_emat = tregs.simulate.generate_emap(rep2_site, fixed=True, fixed_value=1)
```

```python
mutant_bank = []
for i in range(200):
    mutants = np.unique(tregs.wgregseq.mutations_rand(promoter,
                                                rate=0.1,
                                                num_mutants=5000,
                                                number_fixed=True,
                                                keep_wildtype=True))
    mutant_bank.append(mutants)
```

```python
n_NS = len(genome)
ep_wt, er1_wt, er2_wt = -5, -14, -14
e_int = -5
n_p, n_r2 = 5000, 25

l_n_r1 = np.arange(0, 50, 2)

signal_ratio_AND = []
signal_r1_AND = []
signal_r2_AND = []

for n_r1 in l_n_r1:
    r1 = []
    r2 = []
    for i in range(200):
        df = tregs.simulate.sim(promoter, tregs.simulate.doublerep_pbound,
                                [rnap_site, rep1_site, rep2_site],
                                *[n_NS, n_p, n_r1, n_r2, rnap_emat, r1_emat, r2_emat,
                                  ep_wt, er1_wt, er2_wt, e_int, 'AND'],
                                preset_mutants=mutant_bank[i],
                                scaling_factor=10**6)

        mut_list = df['seq'].values
        mu_data = df['norm_ct_1']
        nbins = 2
        upper_bound = np.median(mu_data)

        footprint = tregs.footprint.get_info_footprint(mut_list, mu_data, promoter, nbins, upper_bound,
                                                    pseudocount=10**(-6))
        r1.append(np.mean(footprint[(115):(115+10)]))
        r2.append(np.mean(footprint[(115+20):(115+30)]))
    r1 = np.asarray(r1)
    r2 = np.asarray(r2)
    signal_r1_AND.append((np.mean(r1), np.std(r1)))
    signal_r2_AND.append((np.mean(r2), np.std(r2)))
    signal_ratio_AND.append((np.mean(r1 / r2), np.std(r1 / r2)))
```

```python
n_NS = len(genome)
ep_wt, er1_wt, er2_wt = -5, -14, -14
e_int = -5
n_p, n_r2 = 5000, 25

l_n_r1 = np.arange(0, 50, 2)

signal_ratio_OR = []
signal_r1_OR = []
signal_r2_OR = []

for n_r1 in l_n_r1:
    r1 = []
    r2 = []
    for i in range(200):
        df = tregs.simulate.sim(promoter, tregs.simulate.doublerep_pbound,
                                [rnap_site, rep1_site, rep2_site],
                                *[n_NS, n_p, n_r1, n_r2, rnap_emat, r1_emat, r2_emat,
                                  ep_wt, er1_wt, er2_wt, e_int, 'OR'],
                                preset_mutants=mutant_bank[i],
                                scaling_factor=10**6)

        mut_list = df['seq'].values
        mu_data = df['norm_ct_1']
        nbins = 2
        upper_bound = np.median(mu_data)

        footprint = tregs.footprint.get_info_footprint(mut_list, mu_data, promoter, nbins, upper_bound,
                                                    pseudocount=10**(-6))
        r1.append(np.mean(footprint[(115):(115+10)]))
        r2.append(np.mean(footprint[(115+20):(115+30)]))
    r1 = np.asarray(r1)
    r2 = np.asarray(r2)
    signal_r1_OR.append((np.mean(r1), np.std(r1)))
    signal_r2_OR.append((np.mean(r2), np.std(r2)))
    signal_ratio_OR.append((np.mean(r1 / r2), np.std(r1 / r2)))
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.axhline(y=1, color='black', linestyle='--', alpha=0.5)
ax.axvline(x=25, color='black', linestyle='--', alpha=0.5)
ax.fill_between(l_n_r1,
                 np.asarray(signal_ratio_OR)[:, 0] - np.asarray(signal_ratio_OR)[:, 1],
                 np.asarray(signal_ratio_OR)[:, 0] + np.asarray(signal_ratio_OR)[:, 1],
                 color='#AB85AC', alpha=0.25)
ax.scatter(l_n_r1, np.asarray(signal_ratio_OR)[:, 0], label='Second repressor binding site', color='#AB85AC', s=20)

ax.set_xlabel('Copy number of the first repressor', fontsize=13)
ax.set_ylabel('Ratio of average mutual\ninformation between two\nrepressor binding sites', fontsize=13)

plt.tight_layout()
plt.savefig('../rebuttal_plots/figR6.pdf', dpi=300)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(4,8))
ax1, ax2 = axes

ax1.fill_between(l_n_r1,
                 np.asarray(signal_r1_AND)[:, 0] - np.asarray(signal_r1_AND)[:, 1],
                 np.asarray(signal_r1_AND)[:, 0] + np.asarray(signal_r1_AND)[:, 1],
                 color='#7AA974', alpha=0.25)
ax1.scatter(l_n_r1, np.asarray(signal_r1_AND)[:, 0], label='First repressor binding site', color='#7AA974', s=20)

ax1.fill_between(l_n_r1,
                 np.asarray(signal_r2_AND)[:, 0] - np.asarray(signal_r2_AND)[:, 1],
                 np.asarray(signal_r2_AND)[:, 0] + np.asarray(signal_r2_AND)[:, 1],
                 color='#AB85AC', alpha=0.25)
ax1.scatter(l_n_r1, np.asarray(signal_r2_AND)[:, 0], label='Second repressor binding site', color='#AB85AC', s=20)
ax1.set_ylabel('Average mutual information (bits)', fontsize=13)
ax1.set_title('AND logic', fontsize=15)
ax1.set_yticks(np.arange(0, 0.015, 0.002))

ax2.fill_between(l_n_r1,
                 np.asarray(signal_r1_OR)[:, 0] - np.asarray(signal_r1_OR)[:, 1],
                 np.asarray(signal_r1_OR)[:, 0] + np.asarray(signal_r1_OR)[:, 1],
                 color='#7AA974', alpha=0.25)
ax2.scatter(l_n_r1, np.asarray(signal_r1_OR)[:, 0], label='First repressor binding site', color='#7AA974', s=20)

ax2.fill_between(l_n_r1,
                 np.asarray(signal_r2_OR)[:, 0] - np.asarray(signal_r2_OR)[:, 1],
                 np.asarray(signal_r2_OR)[:, 0] + np.asarray(signal_r2_OR)[:, 1],
                 color='#AB85AC', alpha=0.25)
ax2.scatter(l_n_r1, np.asarray(signal_r2_OR)[:, 0], label='Second repressor binding site', color='#AB85AC', s=20)
ax2.set_xlabel('Copy number of the first repressor', fontsize=13)
ax2.set_ylabel('Average mutual information (bits)', fontsize=13)
ax2.set_title('OR logic', fontsize=15)
#ax2.set_ylim(0.008, 0.013)
ax2.legend(bbox_to_anchor=(1, -0.2), frameon=False, fontsize=13)

plt.tight_layout()
plt.savefig('../rebuttal_plots/figR7.pdf', dpi=300)
```
