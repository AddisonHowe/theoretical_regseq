```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def boltzmann_ratio(R1, R2, P, e1, e2, ep, w,
                    mut_e1=0, mut_e2=0):
    numerator = R1 * R2 * np.exp(- (e1 + mut_e1) - (e2 + mut_e2) - w)
    denominator = P * np.exp(- ep)
    denominator += R1 * P * np.exp(- (e1 + mut_e1) - ep)
    denominator += R2 * P * np.exp(- (e2 + mut_e2) - ep)
    return numerator / denominator
```

```python
R1 = np.arange(0, 51, 1)
R2 = 25
P = 5000
e1 = -14
e2 = -14
ep = -5
w = -5
```

```python
res_R1 = boltzmann_ratio(R1, R2, P, e1, e2, ep, w, mut_e1 = 2)
res_R2 = boltzmann_ratio(R1, R2, P, e1, e2, ep, w, mut_e2 = 2)

fig, ax = plt.subplots()
ax.plot(R1, res_R1, label='mutation in R1')
ax.plot(R1, res_R2, label='mutation in R2')
ax.set_xlabel('R1')
ax.set_ylabel('R1 + R2 / P')
ax.legend()
```
